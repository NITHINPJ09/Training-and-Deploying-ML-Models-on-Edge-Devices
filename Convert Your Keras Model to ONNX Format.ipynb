```
import onnx
import tf2onnx
import tensorflow
```

```
model = tensorflow.keras.models.load_model('/content/drive/My Drive/Colab Notebooks/cifar_classifier.h5')
onnx_model, _ = tf2onnx.convert.from_keras(model)
onnx.save(onnx_model, '/content/drive/My Drive/Colab Notebooks/cifar_classifier.onnx')
```

```
input_onnx_model = '/content/drive/My Drive/Colab Notebooks/cifar_classifier.onnx'
model = onnx.load(input_onnx_model)
onnx.checker.check_model(model)
print('The ONNX model is checked!')
```
