```
import onnxruntime
import numpy as np
import cv2
```

```
input_onnx_model = '/content/drive/My Drive/Colab Notebooks/cifar_classifier.onnx'
sess = onnxruntime.InferenceSession(input_onnx_model)
```

```
input_name = sess.get_inputs()[0].name
print('input name', input_name)
input_shape = sess.get_inputs()[0].shape
print('input shape', input_shape)
input_type = sess.get_inputs()[0].type
print('input type', input_type)
```

```
output_name = sess.get_outputs()[0].name
print('output name', output_name)
output_shape = sess.get_outputs()[0].shape
print('output shape', output_shape)
output_type = sess.get_outputs()[0].type
print('output type', output_type)
```

```
img = cv2.imread('/content/drive/My Drive/Colab Notebooks/horse.jpg')
img = cv2.resize(img,(32,32))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = np.array([img]).astype('float32') / 255.0
```

```
res = sess.run([output_name], {input_name: img})
predicted_index = np.argmax(res[0])
predicted_probability = np.max(res[0])
label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'] 
label_names.sort()
predicted_label =  label_names[predicted_index]
print('Predicted: {}, Probability {:.2f}%'.format(predicted_label, predicted_probability*100))
```
